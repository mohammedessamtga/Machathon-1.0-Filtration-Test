```
import pandas as pd 
import numpy as np 

data = pd.read_csv("/content/train.csv")
```

```
data.describe()
```

```
data.drop(['Unnamed: 0','WellID'], axis=1,inplace=True)
```

```
Y = data['TotalOilInNext6Months']
data.drop('TotalOilInNext6Months',axis=1,inplace=True)
np.shape(Y)
```

```
X = data 
```

```
X
```

# Neural Network 

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```
import keras 
from tensorflow.keras import regularizers
from keras import layers
model = keras.Sequential(
    [
        #layers.Dense(784, activation="relu", name="layer1",kernel_regularizer=regularizers.l2(1e-4)),
        #layers.Dense(256, activation="relu", name="layer2",kernel_regularizer=regularizers.l2(1e-4)),
        layers.Dense(128, activation="relu", name="layer3",kernel_regularizer=regularizers.l2(1e-4)),
        layers.Dense(68, activation="relu", name="layer4",kernel_regularizer=regularizers.l2(1e-4)),
        layers.Dense(32, activation="relu", name="layer5",kernel_regularizer=regularizers.l2(1e-4)),
        layers.Dense(1, activation="relu", name="layer6",kernel_regularizer=regularizers.l2(1e-4))
    ]
)
```

```
model.compile(loss='mean_squared_error', optimizer='adam')
```

```
X_train = np.array(X_train)
X_test  = np.array(X_test)
y_train = np.array(y_train)
y_test  = np.array(y_test)
X  = np.array(X)
Y  = np.array(Y)
```

```
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100)
```

```
model.fit(X,Y,epochs=50)
```

```
test = pd.read_csv('/content/test.csv')
```
